```python
#CONSTRUCTOR AND MAGIC METHODS, CREATING CUSTOM DATA TYPE IN PYTHON

#non-parameterized constructor
class Atm:
  # 1st Magic method; Constructor; has super power of getting executed without being called directly,
  #it is triggered automatically when the object of the class is created)
  def __init__(self):
    self.pin= ''
    self.balance=0
    self.menu()

  def menu(self):
    user_input = input("""
    Hi how can I help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    Anything else to exit.
    """)

    if user_input == '1':
      #create_pin
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_bal()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()


  def create_pin(self):
    user_pin = input('enter your pin: ')
    self.pin = user_pin

    user_balance = int(input('input your balance: '))
    self.balance = user_balance

    print('pin created successfully')
    self.menu()

  def change_pin(self):
    old_pin = input("enter old pin")

    if old_pin == self.pin:
      self.pin == input("enter new pin")
      print("pin successfully changed")
      self.menu()
    else:
      print("ni karne de sakta re baba")
      self.menu()

  def check_bal(self):
    user_pin = input("enter pin: ")

    if user_pin == self.pin:
      print(f"Your balance is {self.balance}")
    else:
      print('chal nikal yaha se')
    self.menu()

  def withdraw(self):
    withdrawal_amt = int(input("enter amount to withdraw: "))
    pin = input("enter your pin")
    if pin == self.pin:
      if withdrawal_amt <= int(self.balance):
        self.balance = (int(self.balance) - withdrawal_amt)
        print(f"Amount withdwarn successfully. Current balance {int(self.balance)}")
      else:
        print("chal nikal garib saale")
    self.menu()

```

```python
#CREATING CUSTOM DATATYPE IN PYTHON
#parameterized constructor
#needs parameter while creating an object

class Fraction:
  def __init__(self,x,y):
    self.num = x
    self.den = y
#2nd magic method __str__ ; triggered by 'print' to display datatype in desired format
  def __str__(self):
    return '{}/{}'.format(self.num, self.den)

#3rd magic method __add__ ; triggered by '+' operator
  def __add__(self,other):
    new_num = (self.num*other.den + other.num*self.den)
    new_den = (self.den * other.den)
    return '{}/{}'.format(new_num, new_den)

#4th magic method __sub__ ; triggered by '-' operator
  def __sub__(self,other):
    new_num = (self.num*other.den - other.num*self.den)
    new_den = (self.den * other.den)
    return '{}/{}'.format(new_num, new_den)

#5th magic method __mul__ ; triggered by '*' operator
  def __mul__(self,other):
    new_num = (self.num*other.num)
    new_den = (self.den * other.den)
    return '{}/{}'.format(new_num, new_den)

#6th magic method __divide__ ; triggered by '/' operator
  def __truediv__(self,other):
    new_num = (self.num * other.den)
    new_den = (self.den * other.num)
    return '{}/{}'.format(new_num, new_den)

#calling non-magic methods
  def convert_to_decimal(self):
    return self.num/self.den
```

```python
#needs parameter while creating an object
fr1 = Fraction(2,3)
fr2 = Fraction(1,2)
```

```python
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```
